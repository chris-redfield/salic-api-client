```python
import requests
```

```python
url = "http://api.salic.cultura.gov.br/v1/projetos/?limit=100&format=csv&sort=PRONAC&offset="
offset = 0

# aumentando o numero de projetos para adicionar uma ultima iteracao
num_projetos = 89900
```

## Testing the request

```python
response = requests.get(url + str(offset))
```

```python
response.text
```

## iterating the offset

```python
with open('projetos.csv', 'a') as the_file:
    for x in range(int(num_projetos/100)):
        response = requests.get(url + str(offset))
        if(offset > 0):
            the_file.write(response.text.split("\n",1)[1])
        else:
            the_file.write(response.text)
        offset += 100
```
